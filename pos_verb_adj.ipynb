```python
#Write a Python program that takes a paragraph of text, performs part-of-speech tagging, and then:
input_para="The experiment seemed straightforward and there were plenty of scientists willing to try it. Many did. It was wonderful to have a simple laboratory experiment on fusion to try after the decades of embarrassing attempts to control hot fusion. This effort required multi-billion dollar machines whose every success seemed to be capped with an unanticipated failure. ‘Cold fusion’ seemed to provide, as Martin Fleischmann said during the course of that famous Utah press conference, ‘another route’—the route of little science."
import re
para= re.sub(r'[^\w\s.]', '', input_para)
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
multi_sent=sent_tokenize(para)
multi_words=word_tokenize(para)
pos_words=nltk.pos_tag(multi_words)
print(pos_words)

print("\nTokenising words from muliple sentences\n")
for sentindex,sent in enumerate(multi_sent):
    words=word_tokenize(sent) #tokenising the words
    pos=nltk.pos_tag(words)
    print(sentindex,pos)

#Prints a sorted list of all unique verbs in the text.
verbs=[]
i=0
for p in pos_words:
    if p[1].startswith('VB'):
        i+=1
        print("Verb ",i,p[0].lower())
        if p[0] not in verbs:
            verbs.append(p[0].lower())
verbs.sort()
print(f"\nList of sorted verbs: \n {verbs}\n")

#Prints a sorted list of all unique adjectives in the text.
adjs=[]
i=0
for p in pos_words:
    if p[1].startswith('JJ'):
        i+=1
        print("ADJECTIVE ",i,p[0])
        if p[0] not in adjs:
            adjs.append(p[0])
adjs.sort()
print(f"\nList of sorted adjectives: \n {adjs}\n")

#Counts the total number of verbs and adjectives.
print("Total number of verbs and adjectives",len(verbs)+len(adjs),"\n")
```
