```python
import nltk
nltk.download('punkt_tab')
nltk.download('brown')
from collections import defaultdict
import re
with open ('samplefile.txt','r') as f:
    content=f.read()
print("FILE CONTENT\n",content)

from textblob import TextBlob
tb=TextBlob(content)
sent=tb.sentences
print("\nSENTENCES")
for s in sent:
    print(sent.index(s),s)
type(sent)
allwords=tb.words
noun_phrases=tb.np_counts

#removing stopwords
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words=set(stopwords.words('english'))
words=[w for w in allwords if w not in stop_words]
#finding word frequencies
word_freq=defaultdict(int)
for w in words:
    cleaned_w=re.sub(r'[^a-zA-Z]','',w.lower())
    if cleaned_w and cleaned_w not in noun_phrases:
        word_freq[cleaned_w]+=1
#finding sentence score
sent_score=defaultdict(int)
for i,s in enumerate(sent):
    for wrd in s.words:
        cleaned_wrd=re.sub(r'[^a-zA-Z]','',wrd.lower())
        if cleaned_wrd in word_freq:
            sent_score[i]+=word_freq[cleaned_wrd]
print("\nSENTSCORE")
for k,v in sent_score.items():
    print(k,v)

#sorting sent_score dictionary according to the sentence score
sorted_sent_score=sorted(sent_score.items(),key=lambda x:x[1],reverse=True)
print("\nSORTED SENTENCE SCORE")
print(sorted_sent_score)

#selecting 5 sentences with best sentence score
best_score_sent_index=[s[0] for s in sorted_sent_score[:5]]
best_score_sent=[str(sent[i]) for i in best_score_sent_index]
print("\nTOP 5 SENTENCES WITH BEST SENTENCE SCORE")
for i,s in zip(sorted_sent_score[:5],best_score_sent):
    print(f"Sentence Score:{i[1]}, Sentence no: {i[0]}")
    print(f"{s}\n")

print("\nSUMMARY")
print(" ".join([str(sent[i]) for i in sorted(best_score_sent_index)]))




```
